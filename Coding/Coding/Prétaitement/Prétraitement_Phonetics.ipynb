```
# ouvrir un fichier pickle
def open_pickle(path):
  with open(path, 'rb') as handle: 
    doc= pickle.load(handle)
    return doc

# sauvgarder un fichier pickle
def save_pickle(variable,filename):
    with open(filename, 'wb') as handle:
         pickle.dump(variable, handle, protocol=pickle.HIGHEST_PROTOCOL)

# toknization
def tokenization(text):
    return word_tokenize(text)

# verifier si un mot est arabizi
def is_arabizi(word):
  if len(re.findall(r'^[a-zA-Z0-9.]+$',word)) > 0:
    return 1
  else: 
    return 0

# remplacer chaque chiffre avec son alphabet
def keep_original(word):
  word = re.sub("5", "kh", word)
  word = re.sub("3", "a", word)
  word = re.sub("9", "k", word)
  word = re.sub("7", "h", word)
  word = re.sub("6", "t", word)
  word = re.sub("2", "ou", word)
  return word

# retourne la clé phonétique du mot
def phonetic(word):
  #delete repiting letters
  return phonetics.metaphone(word)
```

# Suppression des chiffres isolé


`Le but est de s'assuré qu'aucun chiffre isolé existe dans le corpus, c'est une sorte de nétoyage finale du corpu. nous allons supprimer les chiffre qui n'ont pas des alphabet pour les remplacer.`

```
def remove_number(text):
   puncts = ['0','2','4','8' ] #ces chiffres n'ont pas des alphabet pour la prononciation en arabizi.
   x = str(text)
   for punct in puncts:
      if punct in x:
          x = x.replace(punct,'')
   return x
print(remove_number('gello38 0aje'))   
```

```
for indice in donnee.keys(): 
    text=str(donnee[indice]['text'])
    text1=remove_number(text)
    donnee[indice]['text']=text1
```

# Normalisation des mots arabizi

>le but est d'assurer le bon fonctionnement de l'algorithme phonitiques, car nous avons constaté des cas d'erreur dans la manipulation des mots comportons des chiffres.(présence des faux negatif). donc pour chaque mots arabizi nous unifions ces lettres

>nous normalisons les mots ainsi la valeurs originale du mots seras sauvgarder dans un dictionnaire pour restituer sa valeurs à la fin de l'algorithme



```
origine_of_word=defaultdict()  
for indice in donnee.keys(): 
  for word in  tokenization(donnee[indice]['text']):       
    if( is_arabizi(word) == 1):
      key=keep_original(word)
      origine_of_word[key]=word
origine_of_word
```

# Traitement phonitique

`étape 1: regroupement phonitique `

> nous mesurons la valeurs phonitique pour chaque mots arabizi 

> nous tarvaillons avec des dictionnaires de fréquence, pour garder la fréquence des mots dans le dictionnaire qui seras utilisé dans les étapes suivantes

```
dic= ConditionalFreqDist()      # { phonetics key: liste of words that have the same key phonectis avec leurs fréquences d'apparition dans le corpus}

for word in origine_of_word.keys():     #sachant que instance est une liste qui contient les mots tocknisé
    key=phonetic(word)
    dic[key][word]+=1
dic
```

```
for keys in dic['LFL']:
  print(keys)
```

`étape 2: groupement de similarité`

> on remplace les mots du dictionnaire par la valeurs la plus fréquente 


```
existe=[] # contien tous les mots manipulé ie les mots ayant une clé phonitique pour ne pas recalculer deux fois en cas d'erreur

phonetic_of_word=defaultdict()  # { word: its phonetics key } le dictionnaire ayant comme clé le mots arabizi originale et comme valeur: le mots phonitique à remplacer (le mots le plus fréquent de la liste)

for key in dic:  
  print("__________________________________")
  while(len(dic[key].keys())!= 0):  
    print(key," :: il reste",list(dic[key])) 
    max_val= dic[key].most_common(1)[0][0]                  # 1 on extrais le mots le plus fréquent
    if (max_val not in existe):     # pour éviter toute erreur: si un mots est déja manipulé on ne modifie pas une deuxième fois valeurs phonitique
      phonetic_of_word[max_val]=origine_of_word[max_val] # le mots le plus fréquent est replacer par sa valeurs originale
      dic[key].pop(max_val) # on supprime le mots le plus fréquent du dictionnaire
      existe.append(max_val) # on l'ajoute à la liste des mots traité
      for word in list(dic[key]):
       #print(word)       
          if(word not in existe and distance(max_val,word)<3):
            print('word  ',word, ' replaced by  ',max_val)
            phonetic_of_word[word]=origine_of_word[max_val] # si la condition satisfaiste alors on cree une nouvel clé du dictionnaire = le mots arabizi et sa valeur= le mots le plus fréquent
            existe.append(word) # ajouter à la liste des mots traité 
            dic[key].pop(word) # supprimer du dictionnaire phonitique
            
```

# Remplacer les mots arabizi dans chaque instancepar la valeurs valeurs équivalente

```
donnee_phonetics=defaultdict()
for indice in donnee.keys():
  sentence=donnee[indice]['text'] 
  for word in tokenization(donnee[indice]['text']):
    if( is_arabizi(word) ):
      sentence=sentence.replace(word,phonetic_of_word[keep_original(word)]) #phonetic_of_word[word] gives the key phonetics without reexicute the function phonetics
      print(word,' ==== >',phonetic_of_word[keep_original(word)])
  print(indice,' :: ',donnee[indice]['text']," ===> ",sentence)
  donnee_phonetics[indice]={'text': sentence, 'label': donnee[indice]['label'] }
```
