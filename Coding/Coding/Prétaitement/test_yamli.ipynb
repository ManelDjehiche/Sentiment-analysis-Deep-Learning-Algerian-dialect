```
#télécharger les librarys
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
import numpy as np
import pandas as pd
import nltk
from nltk import re
from nltk import defaultdict
import io
from nltk import  word_tokenize
from nltk import  sent_tokenize
from nltk.corpus import stopwords
import codecs
import csv
import pickle
from gensim.parsing.preprocessing import strip_short
import nltk
from nltk.probability import ConditionalFreqDist
import time
```

```
# ouvrir et sauvgarder un fichier pickle
def open_pickle(path):
  with open(path, 'rb') as handle: 
    doc= pickle.load(handle)
    return doc

def save_pickle(variable,filename):
    with open(filename, 'wb') as handle:
         pickle.dump(variable, handle, protocol=pickle.HIGHEST_PROTOCOL)

#return la tokenisation d'une chaine de caractère 
def tokenization(text):
    return word_tokenize(text)

#verifier si le mot est arabizi, la fonction retourne vrai, sinon elle retourne faux
def is_arabizi(word):
  if len(re.findall(r'^[a-zA-Z0-9]+$',word)) > 0:
    return 1
  else: 
    return 0
```

```
#vérification, la fonction retourne vrai si le mots ne contient aucun lettre arabe, que des lettres latins et des chiffres
print(is_arabizi("fkfhfk55ىلاتلhghghuijkj"))
print(is_arabizi("fىلhاتل"))
print(is_arabizi("ggjgjgj554gg"))
print(is_arabizi("fىلhاتلghghghghgh552 ghgh"))
print(is_arabizi("555kjkjkkjkjkj"))
print(is_arabizi("بالبتنلاب يتتبتبتب"))
```

```
def tansliteration(myInput_ymli): #reçoit le mot arabizi, et retourne sa translétiration
      options = webdriver.ChromeOptions()
      options.add_argument('--headless')
      options.add_argument('--no-sandbox')
      options.add_argument('--disable-dev-shm-usage')
      # open it, go to a website, and get results
      driver_ymli = webdriver.Chrome('chromedriver',options=options)
      try:
        driver_ymli.get("https://www.yamli.com/fr/") 
        driver_ymli.find_element_by_xpath('//*[@id="search_input"]').clear()                          # netoyer
        driver_ymli.find_element_by_xpath('//*[@id="search_input"]').send_keys(myInput_ymli.lower())  # Input the Desired text to be translated
      # spelling error making this code not work as expected 
      except NoSuchElementException:  
          print('element input does not found the element')
      time.sleep(1)
      # output_ymli=driver_ymli.find_element_by_xpath('//*[@id="yamliapi_dyn_div"]/div[5]/div/div/div[1]')
      output_ymli=driver_ymli.find_element_by_xpath('//*[@id="yamliapi_dyn_div"]/div[5]/div/div/div[1]/div[2]')
      return output_ymli.text
```
