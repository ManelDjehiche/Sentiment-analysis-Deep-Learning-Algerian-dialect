# construcion de la liste des abréviation

```
# retourne la liste des mots abrégér dans une phrase
def abreviation(text):
  return re.findall(r'\b[^AEIOU_0-9-\u0621-\u064a\ufb50-\ufdff\ufe70-\ufefc\W]+\b', text, flags=re.I)


abreviation('bnjr tous le mondes cv vous allez bn')
```

```
my_list=[]
# extraire les mots abrégé à partir de DT_1
for indice in list(data.index.values): 
    text=str(data['text'][indice])
    res=abreviation(text) #// Avoir la liste de tous les mots abrégés / Pour la construction de la liste
    my_list.extend(res)    
```

```
#2 sauvgarder la liste
data=pd.DataFrame.from_dict(my_list,orient='index')
data.to_csv(r'/content/drive/My Drive/PROJET FIN ETUDE 2019 2020/ressources/abréviation.csv', index = False)

```
