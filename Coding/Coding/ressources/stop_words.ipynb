```
#importer les library
import nltk
from nltk.corpus import stopwords
import re
import pickle
```

```
#ouvrire et sauvragder fichier pickle
def save_pickle(variable,filename):
    with open(filename, 'wb') as handle:
         pickle.dump(variable, handle, protocol=pickle.HIGHEST_PROTOCOL)

def open_pickle(path):
  with open(path, 'rb') as handle: 
    doc= pickle.load(handle)
    return doc
```

```
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

`Traduction de la liste anglais en dialecte algérien.`

```
dz_stopwords=[]
cpt=0
for myInput in en_stopwords:
  cpt=cpt+1
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  # open it, go to a website, and get results
  driver = webdriver.Chrome('chromedriver',options=options)
  # print(driver.page_source)  # results
  driver.get("https://glosbe.com/en/arq/great")                   # Search the URL
  driver.find_element_by_id('wordinput').clear()    
  driver.find_element_by_id('wordinput').send_keys(myInput.lower())  # Input the Desired text to be translated

  driver.find_element_by_id('navigateToWordButton').click()          # Click on the translate button (not Mandatory)

  try:
      print(myInput," ----------- : ",cpt)
      text_output = driver.find_element_by_xpath('//*[@id="phraseTranslation"]/div/div[5]/ul')
      items = text_output.find_elements_by_tag_name("div")
      for item in items:
        text = item.text
        dz_stopwords.append(text)

  except NoSuchElementException:  #spelling error making this code not work as expected
      print('does not found the element')
      pass
```

`Transliteration des mots de stop word algérien écrites en arabizi en lettre arabe.`

```
def is_arabizi(word): #retourn vrai si le mot est arabizi
  return len(re.findall(r'[A-Za-z0-9]+',word))
```
